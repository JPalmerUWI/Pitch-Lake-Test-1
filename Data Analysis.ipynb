# Pitch Lake Test Data Anlysis

## Bookkeeping

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from math import floor
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JPalmerUWI/Pitch-Lake-Test-1/master/WAVE0001.CSV")
```

```python
df = df[["Time [s]","TEMP [C]","HS [V]","PUMP [V]","LAKE [V]"]]
```

```python
df["HS [mV]"] = df["HS [V]"]*1000
df["PUMP [mV]"] = df["PUMP [V]"]*1000
df["LAKE [mV]"] = df["LAKE [V]"]*1000
df["Time [h]"] = (df["Time [s]"]/3600).round(2)
```

```python
df = df[["Time [h]","TEMP [C]","HS [mV]","PUMP [mV]","LAKE [mV]"]]
df
```

```python
x = df["Time [h]"]
y1 = df["HS [mV]"]
y2 = df["PUMP [mV]"]
y3 = df["LAKE [mV]"]

x_ticks = range(0,94,4)

y_ticks = range(-64,30,3)
```

```python
plt.figure(figsize=(25,20), dpi=50)

# Plotting
plt.plot(x,y1, label="High Sulfur", color = 'b')
plt.plot(x,y2, label="Pump", color='r')
plt.plot(x,y3, label="Lake", color='g')

# Styling
# General
plt.legend(fontsize=60)
plt.axhline(y=0, color='k')
plt.grid()

# X axis 
plt.xticks(x_ticks,rotation=45, fontsize=24)
plt.xlabel("Time, hours", fontsize=32, labelpad=20)

# Y axis
plt.yticks(y_ticks, fontsize=24)
plt.ylabel("Voltage, mV", fontsize=32, labelpad=20)


plt.show()
```
